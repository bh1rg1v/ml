```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
```

```python
# loading dataset

from sklearn.datasets import load_iris

iris = load_iris()
```

```python
iris_data = pd.DataFrame(iris.data)
```

```python
iris_data.columns = iris.feature_names
```

```python
iris_data
```

```python
# scaling or standardization

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
x_scaled = scaler.fit_transform(iris_data)
```

```python
x_scaled
```

```python
# applying PCA

from sklearn.decomposition import PCA

pca = PCA(n_components=2)
```

```python
pca_scaled = pca.fit_transform(x_scaled)
```

```python
plt.scatter(pca_scaled[:, 0], pca_scaled[:, 1], c=iris.target)
```

```python
# agglomerative clustering, to construct a dendogram

import scipy.cluster.hierarchy as sc

plt.figure(figsize=(20, 7))

sc.dendrogram(sc.linkage(pca_scaled, method="ward"))

plt.title("Dendograms")
plt.xlabel("Sample Index")
plt.ylabel("Eucledian Distance")
```

```python
from sklearn.cluster import AgglomerativeClustering

# model = AgglomerativeClustering(n_clusters=2, affinity="euclidean", linkage="ward")       affinity is depreacted 
model = AgglomerativeClustering(n_clusters=2, linkage="ward")
model.fit(pca_scaled)
```

```python
model.labels_
```

```python
plt.scatter(pca_scaled[:, 0], pca_scaled[:, 1], c=model.labels_)
```

```python
# silhouetter score

from sklearn.metrics import silhouette_score

silhouette_coeffs = []

for k in range(2, 11):

    model = AgglomerativeClustering(n_clusters=k, linkage="ward")
    
    model.fit(x_scaled)

    score = silhouette_score(x_scaled, model.labels_)

    silhouette_coeffs.append(score)
```

```python
# plotting silhouette score

plt.plot(range(2, 11), silhouette_coeffs)

plt.xticks(range(2, 11))

plt.xlabel("Number of Clusters")
plt.ylabel("Silhoutte Coefficient")

plt.show()
```
