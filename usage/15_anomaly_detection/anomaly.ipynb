```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
```

```python
df = pd.read_csv("../../data/healthcare.csv")
df.head()
```

```python
# model

from sklearn.ensemble import IsolationForest

model = IsolationForest(contamination=0.2)

model.fit(df)

pred = model.predict(df)
```
